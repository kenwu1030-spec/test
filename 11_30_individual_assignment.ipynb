```python
# Import part
# use a pipeline as a high-level helper
from transformers import pipeline
from IPython.display import Audio

# Function part
# part1: img2text
def img2text(image_name):
    image_to_text_model = pipeline("image-to-text",
                                   model="Salesforce/blip-image-captioning-base")

    input_image_text = image_to_text_model(image_name)[0]["generated_text"]
    return input_image_text

# part2: text2story
def text2story(text_description):
    story_generation_model = pipeline("text-generation",
                                      model="jtatman/neox_125m_storytelling")
    # Use the passed text_description as the prompt
    generated_story_output = story_generation_model(text_description,
                                                    max_new_tokens=100,
                                                    truncation=True,
                                                    num_return_sequences=1)[0]["generated_text"]
    return generated_story_output

# part3: text2audio
def text2audio(generated_story_output):
    text_to_speech_model = pipeline("text-to-audio",
                                    model="Matthijs/mms-tts-eng")

    output_speech = text_to_speech_model(generated_story_output)
    return output_speech

# Main part
print("Title：Storytelling Application")

# Image Input
image_name = input("Enter an image file name: ")
img_description = img2text(image_name = image_name)
print(f"Generated text: {img_description}")

# Story Generation
generated_story = text2story(text_description = img_description)
print(f"Generated Story: {generated_story}")

# Text-to-Speech Conversion
output_speech = text2audio(generated_story_output = generated_story)
Audio(output_speech['audio'],
      rate=output_speech['sampling_rate'])
```
